## Fetch Text Content from Wikipedia

```python
from wikipediaapi import Wikipedia
wiki = Wikipedia('RAGBot/0.0','en')
doc = wiki.page('Virat_Kohli').text
paragraphs = doc.split('\n\n')
```

```python
import textwrap
```

```python
for i,p in enumerate(paragraphs):
  wrapped_text = textwrap.fill(p,width=100)
  print('------------------------------------------------------------------------')
  print(wrapped_text)
```

## Embeddings

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("Alibaba-NLP/gte-base-en-v1.5", trust_remote_code=True)
```

```python
docs_embed = model.encode(paragraphs,normalize_embeddings=True)
```

```python
docs_embed.shape
```

```python
docs_embed
```

```python
query = 'who is virat kohlis mother'
query_embed = model.encode(query,normalize_embeddings=True)
```

```python
query_embed.shape
```

## Find CLosest Paragraphs to the Query:  

```python
import numpy as np
similarities = np.dot(docs_embed,query_embed.T)  # .T means transpose of the matrix
```

```python
similarities
```

They are the similarity scores

```python
similarities.shape
```

```python
top_3_idx = np.argsort(similarities,axis=0)[-3:][::-1].tolist()
```

```python
top_3_idx
```

```python
most_similar_docs = [paragraphs[idx] for idx in top_3_idx]
```

```python
Context = ''
for i,p in enumerate(most_similar_docs):
  wrapped_text = textwrap.fill(p,width=100)

  print('------------------------------------------------------------------------')
  print(wrapped_text)
  print('------------------------------------------------------------------------')
  Context += wrapped_text + '\n\n'
```

```python
prompt = f'''
use the following Context to asnwer the Question at the end.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context:
{Context}

Question: {query}
'''
```
